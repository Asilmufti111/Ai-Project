```python
from collections.abc import Mapping
from experta import *

class frozendict(Mapping):
  pass

# Facts
class MyHotelGuide(Fact):
  pass


# Rules and knowledge engine
class HotelGuide(KnowledgeEngine):

  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'y'), MyHotelGuide(public_swimming_pool = 'y'), MyHotelGuide(kids_swimming_pool = 'y'), MyHotelGuide(spa_and_gym = 'y'), MyHotelGuide(musical_and_live_preferences = 'n')))
  def joudyan(self):
    print("\n*** The best hotel for you: Joudyan ***")
    return True


  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'y'), MyHotelGuide(public_swimming_pool = 'y'), MyHotelGuide(kids_swimming_pool = 'n'), MyHotelGuide(spa_and_gym = 'y'), MyHotelGuide(musical_and_live_preferences = 'y')))
  def kyona_marina_resort(self):
    print("\n*** The best hotel for you: Kyona Marina Resort ***")
    return True


  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'y'), MyHotelGuide(public_swimming_pool = 'y'), MyHotelGuide(kids_swimming_pool = 'y'), MyHotelGuide(spa_and_gym = 'n'), MyHotelGuide(musical_and_live_preferences = 'n')))
  def shirvan(self):
    print("\n*** The best hotel for you: SHIRVAN Hotel City Yard ***")
    return True


  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'y'), MyHotelGuide(public_swimming_pool = 'n'), MyHotelGuide(kids_swimming_pool = 'n'), MyHotelGuide(spa_and_gym = 'n'), MyHotelGuide(musical_and_live_preferences = 'n')))
  def Movenpick(self):
    print("\n*** The best hotel for you: MÃ¶venpick Al Nawras ***")
    return True


  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'n'), MyHotelGuide(public_swimming_pool = 'n'), MyHotelGuide(kids_swimming_pool = 'n'), MyHotelGuide(spa_and_gym = 'y'), MyHotelGuide(musical_and_live_preferences = 'n')))
  def the_ritz_carlton(self):
    print("\n*** The best hotel for you: The Ritz-Carlton ***")
    return True


  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'n'), MyHotelGuide(public_swimming_pool = 'y'), MyHotelGuide(kids_swimming_pool = 'y'), MyHotelGuide(spa_and_gym = 'y'), MyHotelGuide(musical_and_live_preferences = 'n')))
  def radisson_blu_hotel(self):
    print("\n*** The best hotel for you: Radisson Blu Hotel ***")
    return True


  @Rule(AND (MyHotelGuide(luxury = 'y'), MyHotelGuide(breakfast = 'n'), MyHotelGuide(public_swimming_pool = 'y'), MyHotelGuide(kids_swimming_pool = 'n'), MyHotelGuide(spa_and_gym = 'y'), MyHotelGuide(musical_and_live_preferences = 'n')))
  def Vivienda(self):
    print("\n*** The best hotel for you: Vivienda Hotel Villas ***")
    return True


  @Rule(AND (MyHotelGuide(mid_range = 'y'), MyHotelGuide(breakfast = 'n'), MyHotelGuide(gym = 'n'), MyHotelGuide(kids_club = 'n')))
  def La_Fontaine_Lagoon(self):
    print("\n*** The best hotel for you: La Fontaine Lagoon Resort ***")
    return True


  @Rule(AND (MyHotelGuide(mid_range = 'y'), MyHotelGuide(breakfast = 'y'), MyHotelGuide(gym = 'y'), MyHotelGuide(kids_club = 'n')))
  def Sukoon(self):
    print("\n*** The best hotel for you: Sukoon hotel ***")
    return True


  @Rule(AND (MyHotelGuide(mid_range = 'y'), MyHotelGuide(breakfast = 'y'), MyHotelGuide(gym = 'y'), MyHotelGuide(kids_club = 'y')))
  def Shada(self):
    print("\n*** The best hotel for you: Shada Hotel Al Shati ***")
    return True


  @Rule(AND (MyHotelGuide(mid_range = 'y'), MyHotelGuide(breakfast = 'n'), MyHotelGuide(gym = 'y'), MyHotelGuide(kids_club = 'n')))
  def  Radisson_Blu_Plaza(self):
    print("\n*** The best hotel for you:  Radisson Blu Plaza ***")
    return True


  @Rule(AND (MyHotelGuide(Budget = 'y'), MyHotelGuide(breakfast = 'n')))
  def Areen(self):
    print("\n*** The best hotel for you: Areen Hotel ***")
    return True


  @Rule(AND (MyHotelGuide(Budget = 'y'), MyHotelGuide(breakfast = 'y')))
  def Dora(self):
    print("\n*** The best hotel for you: Dora alroya hotel ***")
    return True


# Welcome Message
print("=====================================================")
print("           Welcome to My Hotel System ")
print("     We will help you decide which hotel is \n  suitable for you to rest based on your answers!")
print("=====================================================\n")
start = input("Are you ready to get started (y/n)? ")
if start=='y'.lower():
  engine = HotelGuide()
  engine.reset()
  luxuryANS = input("\n> Do you Want a luxury hotel (5-7 stars), (y|n)? ").lower()
  if luxuryANS =='y':
    breakfastANS = input ("\n> Do you want the hotel provides breakfast, (y|n)? ").lower()
    public_swimming_poolANS= input ("\n> Do you want the hotel to have a public swimming pool, (y|n)? ").lower()
    kids_swimming_poolANS= input ("\n> Do you want the hotel to have a swimming pool for kids, (y|n)? ").lower()
    spa_and_gymANS= input ("\n> Do you want the hotel to have a spa and gym, (y|n)? ").lower()
    musical_and_live_preferencesANS= input ("\n> Do you want the hotel to have a musical and live preferences, (y|n)? ").lower()

    if (engine.declare (MyHotelGuide(luxury = luxuryANS, breakfast = breakfastANS,
                             public_swimming_pool = public_swimming_poolANS,
                             kids_swimming_pool = kids_swimming_poolANS,
                             spa_and_gym = spa_and_gymANS,
                             musical_and_live_preferences = musical_and_live_preferencesANS)))!= True :
       print("\nSorry, we could not find a hotel for you :(")

    engine.run()

  # mid-range hotel
  else:
     midRangeANS = input("\n> Do you Want a mid range hotel (3-4 stars), (y|n)? ").lower()
     if midRangeANS == 'y':
      breakfastANS = input ("\n> Do you want the hotel provides breakfast, (y|n)? ").lower()
      gymANS = input ("\n> Do you want the hotel to have a gym, (y|n)? ").lower()
      kidsClubANS = input ("\n> Do you want the hotel to have a club for kids, (y|n)? ").lower()

      if (engine.declare(MyHotelGuide(mid_range = midRangeANS,breakfast = breakfastANS, gym = gymANS, kids_club = kidsClubANS)))!= True :
         print("\nSorry, we could not find a hotel for you :(")

      engine.run()

      # budget hotel
     else:
        budgetANS = input("\n> Do you Want a budget hotel (1-2 stars), (y|n)? ").lower()
        if budgetANS == 'y':
          breakfastANS = input ("\n> Do you want the hotel provides breakfast, (y|n)? ").lower()

          if ( engine.declare(MyHotelGuide(Budget = budgetANS, breakfast = breakfastANS)))!= True :
             print("\nSorry, we could not find a hotel for you :(")

          engine.run()

        else:
          print("\nSorry, we could not find a hotel for you :(")

else:
  print ("\nSee you later!")

# a goodbye message
print("Thank you for using My Hotel Guide system :)")
```
